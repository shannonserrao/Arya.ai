**Look at the correlation of the columns**

We see that we have some well correlated features

```python
corr = df.corr()
# corr.style.background_gradient()
```

```python
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=np.bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(21, 19))

# Generate a custom diverging colormap
cmap = "RdBu"

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(
    corr,
    mask=mask,
    cmap=cmap,
    center=0,
    square=True,
    linewidths=0.5,
    cbar_kws={"shrink": 0.5},
)
```

**Plotting the histogram of the features**

```python
cols = ['X'+ str(i) for i in range(1,21)]
fig, ax = plt.subplots(nrows=5,ncols=4, figsize=(40,30))

for ax, col in zip(ax.ravel(), cols):
#     print(ax, col)
    df[df[col]!=0][col].hist(ax=ax, bins=60,label=col)
    ax.set_xlabel(col, fontsize = 30)
```

```python
cols = ['X'+ str(i) for i in range(41,58)]
fig, ax = plt.subplots(nrows=5,ncols=4, figsize=(40,30))

for ax, col in zip(ax.ravel(), cols):
#     print(ax, col)
    df[df[col]!=0][col].hist(ax=ax, bins=60,label=col)
    ax.set_xlabel(col, fontsize = 30)
```
